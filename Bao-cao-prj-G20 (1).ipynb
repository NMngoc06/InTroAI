### I. Thuật toán DFS (Depth - First Search) 

- Thuật toán DFS là kỹ thuật duyệt đồ thị theo chiều sâu, phù hợp duyệt toàn bộ các ô của mê cung. Các nút được xét theo thứ tự độ sâu giảm dần.
- Các ô có giá trị 1 là đường có thể đi được.
- Các ô có giá trị 0 là tường.

#### Mô tả thuật toán:

1. **Khởi tạo**: Đặt ô bắt đầu vào ngăn xếp.
2. **Duyệt**: Lặp lại các bước cho đến khi ngăn xếp rỗng.
    - Lấy ô hiện tại từ đỉnh ngăn xếp.
    - Đánh dấu ô này là đã duyệt.
    - Nếu ô hiện tại là ô kết thúc thì dừng lại và trả về đường đi đã tìm được.
    - Kiểm tra các ô xung quanh. Nếu ô đó chưa được duyệt và có giá trị 1, thêm vào ngăn xếp.
3. **Kết thúc**: Nếu duyệt hết mà không tìm thấy ô kết thúc thì không có đường đi.

#### Cài đặt giải thuật DFS:

```cpp
DFS(N, A, n_0, DICH) {
    fringe ← n_0;
    closed ← Ø;
    while (fringe ≠ Ø) do {
        n ← GET_FIRST(fringe);   // lấy phần tử đầu tiên của fringe
        closed ← closed + n;

        if (n ∈ DICH) then return SOLUTION(n);

        if (Γ(n) ≠ Ø) then fringe ← Γ(n) + fringe;
    }
    return ("No solution");
}
### II. Thuật toán duyệt đồ thị ưu tiên chiều rộng (Breadth-first search - BFS)

Phát triển các nút theo chiều rộng - Các nút được xét theo thứ tự độ sâu tăng dần.

### Mô tả thuật toán:

#### Khởi tạo:
- Chọn ô bắt đầu và ô đích.
- Tạo một hàng đợi (queue) để lưu trữ các ô cần khám phá.
- Tạo một ma trận đã thăm để đánh dấu các ô đã đi qua.

#### Thực hiện tìm kiếm:
- Thêm ô bắt đầu vào hàng đợi.
- Lặp đi lặp lại cho đến khi hàng đợi rỗng hoặc tìm thấy ô đích:
  - Lấy ô đầu tiên trong hàng đợi ra.
  - Kiểm tra xem ô này có phải là ô đích không. Nếu đúng, tìm đường đi và kết thúc.
  - Đánh dấu ô hiện tại là đã thăm.
  - Thêm tất cả các ô kề chưa được thăm và không phải là chướng ngại vật vào hàng đợi.

#### Tìm đường đi:
- Sử dụng một mảng cha để lưu trữ ô trước đó của mỗi ô.
- Khi tìm thấy ô đích, truy ngược lại từ ô đích đến ô bắt đầu theo mảng cha để tìm ra đường đi.

### Cài đặt giải thuật BFS:

```cpp
BFS(N, A, n_0, DICH) {
    fringe ← n_0;
    closed ← Ø;
    while (fringe ≠ Ø) do {
        n ← GET_FIRST(fringe);   // lấy phần tử đầu tiên của fringe
        closed ← closed + n;

        if (n ∈ DICH) then return SOLUTION(n);

        if (Γ(n) ≠ Ø) then fringe ← fringe + Γ(n);
    }
    return ("No solution");
}

## BFS - Các đặc điểm:

- **Tính hoàn chỉnh?** Có (nếu ma trận hữu hạn).
- **Độ phức tạp về thời gian?** \(1 + b + … + b^{(d-1)} = O(b^{(d+1)})\).
- **Độ phức tạp về bộ nhớ?** \(O(b^{(d+1)})\) – Lưu tất cả các node trong bộ nhớ.
- **Tính tối ưu?** Có (nếu chi phí = 1 cho mỗi bước).

### Các ký hiệu được sử dụng trong giải thuật BFS và DFS:
- **fringe:** Cấu trúc kiểu hàng đợi (queue) hoặc ngăn xếp lưu giữ các nút (trạng thái) sẽ được duyệt.
- **closed:** Cấu trúc kiểu hàng đợi (queue) hoặc ngăn xếp lưu giữ các nút (trạng thái) đã được duyệt.
- **G = (N, A):** Cây biểu diễn không gian trạng thái của bài toán.
- **Γ(n):** Tập các trạng thái (nút) con của trạng thái (nút) đang xét n.
- **n_0:** Trạng thái đầu của bài toán (nút gốc của cây).
- **DICH:** Tập các trạng thái đích của bài toán.
- **b:** Hệ số phân nhánh tối đa của cây tìm kiếm.
- **d:** Độ sâu của lời giải có chi phí thấp nhất.

## III. Thuật toán A* search

Tránh việc xét (phát triển) các nhánh tìm kiếm đã xác định (cho đến thời điểm hiện tại) là có chi phí cao.

### Mô tả thuật toán:

#### Khởi tạo:
- Thêm nút bắt đầu vào danh sách mở (open set).
- Khởi tạo các giá trị g(n) và f(n) cho nút bắt đầu.

#### Thực hiện tìm kiếm: Lặp:
- Chọn nút trong danh sách mở có f(n) nhỏ nhất.
- Nếu nút hiện tại là nút đích, trả về đường đi.
- Thêm nút hiện tại vào danh sách đóng (closed set).
- Xét các nút kề của nút hiện tại:
  - Tính toán g(n) và f(n) cho các nút kề.
  - Nếu nút kề chưa nằm trong danh sách mở hoặc đóng, hoặc nếu giá trị g(n) mới nhỏ hơn giá trị cũ, cập nhật lại g(n) và f(n) và thêm nút vào danh sách mở.

#### Tìm đường đi:
- Nếu danh sách mở rỗng, không tìm thấy đường đi.

### Cài đặt giải thuật A*:

```cpp
A*(N, A, n_0, DICH) {
    fringe ← { (n_0, 0) }; // fringe là hàng đợi ưu tiên, chứa nút và chi phí f(n)
    closed ← Ø;
    g ← {}; // g(n): lưu chi phí từ điểm bắt đầu đến n
    g[n_0] ← 0;
    while (fringe ≠ Ø) do {
        n ← GET_FIRST(fringe); // lấy phần tử có f(n) nhỏ nhất trong fringe
        closed ← closed + {n};
        if (n ∈ DICH) then return SOLUTION(n);
        for each neighbor m of n do {
            if m ∉ closed then {
                tentative_g ← g[n] + cost(n, m);
                if m ∉ fringe or tentative_g < g[m] then {
                    g[m] ← tentative_g;
                    f[m] ← g[m] + h(m); // h(m) là hàm heuristic
                    fringe ← fringe + { (m, f[m]) };
                }
            }
        }
    }
    return ("No solution");
}
### Giải thích:
- **fringe:** Hàng đợi ưu tiên, chứa các nút cùng với chi phí tổng hợp \( f(n) = g(n) + h(n) \).
- **g(n):** Chi phí từ điểm bắt đầu đến n.
- **h(n):** Hàm heuristic ước lượng chi phí từ n đến đích.
- **closed:** Tập các nút đã được mở rộng.
- **g:** Từ điển lưu trữ chi phí thực tế từ điểm bắt đầu đến từng nút.

## 1. Chi tiết Phương pháp

### 1.1 Thuật toán A* (A-star)

- **Mục đích**: Tìm đường từ điểm bắt đầu đến điểm kết thúc trong ma trận một cách hiệu quả nhất, sử dụng hàm chi phí tổng hợp. Thuật toán A* là một thuật toán tìm đường tối ưu, kết hợp giữa tìm kiếm theo chiều rộng (BFS) và tìm kiếm thông minh dựa trên ước lượng khoảng cách đến đích.
  
- **Cách thức hoạt động**:
    - **G (Chi phí từ điểm bắt đầu)**: Chi phí đi từ điểm khởi đầu tới điểm hiện tại.
    - **H (Chi phí ước lượng đến đích)**: Ước lượng khoảng cách còn lại từ điểm hiện tại đến điểm kết thúc (sử dụng khoảng cách Euclid).
    - **F (Chi phí tổng cộng)**: Tổng chi phí của một điểm, được tính theo công thức: `f(n) = g(n) + h(n)`.
    - Thuật toán sử dụng **PriorityQueue** (hàng đợi ưu tiên) để chọn ra điểm có chi phí `f` thấp nhất trong mỗi bước. Mỗi khi di chuyển đến điểm mới, thuật toán tính lại chi phí `G` và ước lượng `H`, sau đó cập nhật lại hàng đợi ưu tiên.
    - Khi tìm được đường đến đích, thuật toán đi ngược lại từ điểm kết thúc đến điểm bắt đầu thông qua các điểm cha (`parent`) của mỗi điểm, và đánh dấu đường đi bằng màu sắc.

### 1.2 Thuật toán DFS (Depth-First Search)

- **Mục đích**: Tìm đường từ điểm bắt đầu đến điểm kết thúc bằng cách đi sâu vào các nhánh, sử dụng phương pháp tìm kiếm theo chiều sâu.

- **Cách thức hoạt động**:
    - DFS bắt đầu từ điểm bắt đầu và tiếp tục đi sâu vào các điểm lân cận cho đến khi tìm được điểm kết thúc hoặc không thể đi tiếp. 
    - Thuật toán sử dụng một **stack** (ngăn xếp) để lưu trữ các điểm cần kiểm tra. Mỗi điểm mới được thêm vào ngăn xếp sẽ được khám phá trước, do đó DFS có thể không tìm được đường đi ngắn nhất nếu có nhiều nhánh.
    - Khi đến điểm kết thúc, thuật toán đi ngược lại từ điểm kết thúc đến điểm bắt đầu, đánh dấu đường đi theo thứ tự mà nó đã duyệt qua.

### 1.3 Thuật toán BFS (Breadth-First Search)

- **Mục đích**: Tìm đường ngắn nhất từ điểm bắt đầu đến điểm kết thúc trong ma trận với điều kiện tất cả các ô có cùng chi phí (hoặc không có trọng số).

- **Cách thức hoạt động**:
    - BFS sử dụng **queue** (hàng đợi) để duyệt các điểm theo chiều rộng. Đầu tiên, BFS kiểm tra tất cả các điểm liền kề điểm bắt đầu, sau đó kiểm tra các điểm của các điểm đã duyệt, và tiếp tục cho đến khi tìm được điểm kết thúc.
    - BFS đảm bảo tìm ra đường đi ngắn nhất trong trường hợp ma trận không có chướng ngại vật, vì nó kiểm tra tất cả các điểm ở cùng một mức độ trước khi di chuyển đến mức độ tiếp theo.

## 2. Chi tiết Dữ liệu

### 2.1 Ma trận (Matrix)

- Ma trận trong ứng dụng được mô phỏng dưới dạng một lưới gồm các ô nhỏ, mỗi ô có kích thước 10x10 pixel. Kích thước của ma trận được xác định bởi tham số `size` khi tạo cửa sổ Pygame (ở đây là 1000x1000 pixels).

- **Ma trận trạng thái (`self.E`)**: Ma trận này lưu trữ thông tin về trạng thái của mỗi ô trong ma trận. Mỗi ô có thể có các giá trị:

### 2.2 Lớp Point

- Mỗi ô trong ma trận được đại diện bởi một đối tượng `Point`. Mỗi đối tượng `Point` chứa các thuộc tính:
    - `x, y`: Tọa độ của điểm trong ma trận (theo chỉ số hàng và cột).
    - `G`: Chi phí từ điểm bắt đầu đến điểm hiện tại.
    - `H`: Chi phí ước lượng từ điểm hiện tại đến điểm kết thúc (dựa trên khoảng cách Euclid).
    - `parent`: Con trỏ đến điểm cha của nó, được sử dụng để theo dõi đường đi sau khi tìm được đích.
  
- **So sánh điểm**: Các đối tượng `Point` có phương thức `__gt__` (greater than) để so sánh tổng chi phí `f = G + H` giữa các điểm. Điều này cho phép thuật toán A* chọn ra điểm có chi phí thấp nhất.

### 2.3 Lớp SmallBox

- Mỗi đối tượng `SmallBox` đại diện cho một ô trong ma trận. Lớp này có các thuộc tính:
    - `x, y`: Vị trí của ô trong ma trận.
    - `state`: Trạng thái hiện tại của ô (ví dụ: "empty", "obs", "sou", "dis").
    - `base_color`: Màu sắc của ô, thay đổi theo trạng thái của ô.
  
- **Phương thức `draw`**: Phương thức này vẽ ô lên cửa sổ Pygame với màu sắc tương ứng với trạng thái của ô. Màu sắc giúp phân biệt các trạng thái như điểm bắt đầu, điểm kết thúc, chướng ngại vật, và các ô trên đường đi.

### 2.4 Queue và Stack

- **PriorityQueue** (hàng đợi ưu tiên) được sử dụng trong A* để quản lý các điểm cần khám phá. Điểm có chi phí `f` thấp nhất sẽ được ưu tiên xử lý trước.
- **Stack**: Được sử dụng trong DFS để lưu trữ các điểm cần kiểm tra. Các điểm sẽ được duyệt theo kiểu LIFO (Last In First Out).
- **Deque** (hàng đợi đôi) được sử dụng trong BFS để lưu trữ các điểm cần khám phá. Các điểm được duyệt theo kiểu FIFO (First In First Out).

# Phân Tích và Đánh Giá Kết Quả

## 1. Tổng Quan Về Kết Quả

Sau khi triển khai các thuật toán tìm kiếm A*, DFS và BFS trên ma trận, chúng ta có thể nhận xét về hiệu quả và độ chính xác của từng thuật toán trong việc tìm ra đường đi từ điểm bắt đầu đến điểm kết thúc. Mỗi thuật toán có những ưu điểm và hạn chế riêng, tùy thuộc vào đặc điểm của không gian tìm kiếm và cấu trúc dữ liệu.

### 1.1 Kết quả của Thuật toán A*

Thuật toán **A*** là một thuật toán tìm kiếm tối ưu, nó kết hợp giữa tìm kiếm theo chiều rộng (BFS) và tìm kiếm thông minh dựa trên ước lượng khoảng cách còn lại tới đích. A* luôn tìm được đường đi ngắn nhất nếu có đường đi khả thi.

#### Ưu điểm:
- **Tính tối ưu**: A* tìm đường đi ngắn nhất nhờ sử dụng hàm chi phí tổng hợp (`f = G + H`), trong đó `G` là chi phí từ điểm bắt đầu và `H` là ước lượng khoảng cách còn lại đến điểm kết thúc. Điều này giúp A* không chỉ khám phá những ô gần đích mà còn đảm bảo tính chính xác.
- **Khả năng điều chỉnh**: Việc tính toán hàm chi phí có thể điều chỉnh cho phù hợp với các tình huống khác nhau, chẳng hạn như sử dụng các trọng số khác nhau cho các hướng di chuyển hoặc các loại chướng ngại vật khác nhau.

#### Hạn chế:
- **Hiệu suất phụ thuộc vào độ phức tạp của không gian tìm kiếm**: A* có thể trở nên chậm nếu không gian tìm kiếm quá lớn hoặc có quá nhiều chướng ngại vật.
- **Tính toán yêu cầu tài nguyên cao**: A* sử dụng PriorityQueue, và việc tính toán `H` có thể tiêu tốn tài nguyên nếu không được tối ưu tốt.

### 1.2 Kết quả của Thuật toán DFS

Thuật toán **DFS** tìm đường đi bằng cách đi sâu vào các nhánh mà không quay lại cho đến khi không thể đi tiếp hoặc đạt đích.

#### Ưu điểm:
- **Dễ triển khai**: DFS rất đơn giản để triển khai và có thể áp dụng cho nhiều loại bài toán khác nhau.
- **Khám phá nhanh các nhánh**: Nếu đích nằm gần một nhánh nào đó, DFS sẽ phát hiện rất nhanh mà không cần phải kiểm tra nhiều nhánh khác.

#### Hạn chế:
- **Không đảm bảo tìm đường ngắn nhất**: Vì DFS đi sâu vào các nhánh mà không kiểm tra các nhánh khác, nên nó có thể tìm ra các đường đi không tối ưu, đặc biệt là trong trường hợp có nhiều chướng ngại vật hoặc điểm đích nằm xa điểm bắt đầu.
- **Hiệu suất kém trong không gian tìm kiếm phức tạp**: Nếu không gian tìm kiếm quá rộng hoặc có nhiều nhánh, DFS có thể tốn rất nhiều thời gian và không gian bộ nhớ, vì nó có thể duyệt qua các nhánh không cần thiết.

### 1.3 Kết quả của Thuật toán BFS

Thuật toán **BFS** là thuật toán tìm kiếm theo chiều rộng, đảm bảo tìm ra đường đi ngắn nhất trong không gian tìm kiếm, miễn là tất cả các ô có cùng trọng số hoặc không có trọng số.

#### Ưu điểm:
- **Tính tối ưu**: BFS tìm ra đường đi ngắn nhất nếu không gian tìm kiếm không có trọng số. Nó đảm bảo rằng đường đi được tìm ra có số bước đi ít nhất.
- **Đảm bảo tính hoàn chỉnh**: BFS đảm bảo sẽ tìm được đường đi nếu có một đường đi khả thi, vì nó duyệt toàn bộ không gian tìm kiếm một cách hệ thống.

#### Hạn chế:
- **Hiệu suất kém với không gian tìm kiếm lớn**: BFS phải duyệt qua tất cả các điểm ở mỗi mức độ (level) trước khi chuyển sang mức độ tiếp theo, điều này có thể gây tốn tài nguyên, đặc biệt là khi không gian tìm kiếm có quá nhiều điểm và chướng ngại vật.
- **Không hiệu quả với không gian tìm kiếm có trọng số**: BFS không tối ưu nếu có trọng số, vì nó không xét đến chi phí hoặc độ khó của các ô trong quá trình tìm kiếm.

## 2. So Sánh Hiệu Quả Của Các Thuật Toán

### 2.1 Thời gian thực thi

- **A***: Thời gian thực thi của A* phụ thuộc vào độ phức tạp của không gian tìm kiếm, nhưng nó nhanh hơn BFS trong nhiều trường hợp vì A* chỉ duyệt các ô có thể đi và ước lượng khoảng cách tới đích.
- **DFS**: DFS có thể nhanh trong những tình huống đơn giản, nhưng trong trường hợp không gian tìm kiếm phức tạp hoặc có nhiều nhánh, thời gian thực thi có thể rất lâu.
- **BFS**: Thời gian thực thi của BFS sẽ dài nếu không gian tìm kiếm rộng hoặc chứa nhiều chướng ngại vật, bởi BFS cần kiểm tra tất cả các điểm cùng một mức độ trước khi chuyển sang mức độ tiếp theo.

### 2.2 Độ chính xác và độ tối ưu

- **A***: Đảm bảo tối ưu về đường đi, vì nó luôn tìm ra đường đi ngắn nhất nếu có.
- **DFS**: Không đảm bảo tối ưu, vì nó có thể đi qua những nhánh không cần thiết và không phải lúc nào cũng tìm ra đường đi ngắn nhất.
- **BFS**: Đảm bảo tối ưu nếu tất cả các ô có cùng trọng số, vì nó luôn tìm ra đường đi ngắn nhất.

### 2.3 Khả năng xử lý các chướng ngại vật

- **A***: Có khả năng xử lý chướng ngại vật tốt, vì nó có thể tính toán lại hướng đi khi gặp phải chướng ngại vật và ưu tiên đi những con đường có chi phí thấp hơn.
- **DFS**: Có thể xử lý chướng ngại vật nhưng nếu gặp phải nhiều chướng ngại vật, DFS có thể sẽ duyệt qua rất nhiều nhánh không cần thiết, làm giảm hiệu quả.
- **BFS**: Xử lý tốt các chướng ngại vật, vì nó duyệt toàn bộ không gian tìm kiếm theo từng mức độ, giúp tránh các khu vực bị chắn.

## 3. Kết Luận

Mỗi thuật toán đều có những ưu điểm và nhược điểm riêng:

- **A*** là lựa chọn tốt nhất khi cần tìm đường đi ngắn nhất và không gian tìm kiếm có thể được đánh giá bằng cách tính toán chi phí. Đây là thuật toán tối ưu cho các bài toán tìm kiếm đường đi trong không gian có trọng số hoặc chướng ngại vật.
  
- **DFS** có thể thích hợp cho các bài toán đơn giản nhưng không đảm bảo tìm đường ngắn nhất trong không gian phức tạp.

- **BFS** là thuật toán thích hợp khi không gian tìm kiếm không có trọng số và đảm bảo tính tối ưu nhưng có thể kém hiệu quả trong các không gian tìm kiếm lớn.

Tóm lại, nếu yêu cầu về độ tối ưu và hiệu suất là cao, **A*** là sự lựa chọn tốt nhất. Tuy nhiên, nếu cần đơn giản và dễ triển khai, DFS hoặc BFS cũng có thể là sự lựa chọn phù hợp tuỳ thuộc vào yêu cầu của bài toán.

  - Tối ưu hóa hiệu suất của thuật toán A* và tích hợp vào hệ thống.

  - Kiểm tra và xử lý các lỗi trong quá trình thực thi của thuật toán DFS.
- **Tỷ lệ đóng góp**: 25%

### Thành viên 3:

  - Nghiên cứu và triển khai thuật toán BFS (Tìm kiếm theo chiều rộng).
  - Phát triển các chức năng liên quan đến việc vẽ ma trận và điểm bắt đầu, điểm kết thúc.
  - Tích hợp các thuật toán BFS và DFS vào giao diện đồ họa chung.

  - Kiểm tra hiệu quả của các thuật toán, phân tích kết quả và tối ưu hóa các thuật toán.
  - Viết báo cáo, tổng kết và trình bày kết quả cuối cùng.
